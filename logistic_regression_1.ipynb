```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
try:
    df = pd.read_csv("insurance_data.csv")
except FileNotFoundError:
    print("insurance_data.csv not found. Generating dummy data for demonstration.")
    import numpy as np
    ages = np.random.randint(18, 65, 100) # Generate 100 random ages between 18 and 65
    # Simulate insurance buying behavior: older people more likely to buy
    bought_insurance = (ages > 40).astype(int) + np.random.randint(0, 2, 100) # Add some noise
    bought_insurance = np.clip(bought_insurance, 0, 1)
    df = pd.DataFrame({'age': ages, 'bought_insurance': bought_insurance})
df.head()
plt.plot(df.age,df.bought_insurance,marker='+',color='blue')
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
X_test
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```
